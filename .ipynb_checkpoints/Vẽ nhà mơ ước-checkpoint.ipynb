```python

import turtle
pen = turtle.Turtle()
pen.hideturtle()

# đổ mầu phông (cỏ)
pen.getscreen().bgcolor('green')


# đổ màu trời
pen.penup()
pen.goto(-400, -100)
pen.pendown()
pen.color("deepskyblue")
pen.begin_fill()
for i in range(2):
    pen.forward(800)
    pen.left(90)
    pen.forward(500)
    pen.left(90)
pen.end_fill()

# vẽ mặt trời
pen.penup()
pen.goto(-320, 225)
pen.pendown()
pen.color("yellow")
pen.begin_fill()
pen.circle(35)
pen.end_fill()

# vẽ mây
pen.penup()
pen.goto(200, 200)
pen.pendown()
pen.color("white")
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(220, 240)
pen.pendown()
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(230, 215)
pen.pendown()
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(180, 225)
pen.pendown()
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(240, 240)
pen.pendown()
pen.color("white")
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(240, 200)
pen.pendown()
pen.color("white")
pen.begin_fill()
pen.circle(25)
pen.end_fill()

pen.penup()
pen.goto(260, 210)
pen.pendown()
pen.color("white")
pen.begin_fill()
pen.circle(25)
pen.end_fill()

# vẽ nhà
pen.penup()
pen.goto(-100, -100)
pen.pendown()
pen.pensize(3)
pen.color("chocolate", "orange") # (stroke, fill)
pen.begin_fill()
for i in range(4):
    pen.forward(170)
    pen.left(90)
pen.end_fill()

# vẽ ống khói
pen.penup()
pen.goto(20, 130)
pen.pendown()
pen.color("brown", "firebrick")
pen.begin_fill()
for i in range(2):
    pen.forward(40)
    pen.left(90)
    pen.forward(100)
    pen.left(90)
pen.end_fill()

# vẽ mái nhà
pen.penup()
pen.goto(-127, 70)
pen.pendown()
pen.begin_fill()
for i in range(3):
    pen.forward(225)
    pen.left(120)
pen.end_fill()

# vẽ cửa sổ 01
pen.penup()
pen.goto(0, 0)
pen.pendown()
pen.color("black", "white")
pen.begin_fill()
for i in range(4):
    pen.forward(50)
    pen.left(90)
pen.end_fill()

# Vẽ đường ngang cửa sổ
pen.penup()
pen.goto(0, 25)
pen.pendown()
pen.color("black")
pen.forward(50)

# Vẽ đường dọc cửa sổ
pen.penup()
pen.goto(25, 0)
pen.pendown()
pen.left(90)
pen.forward(50)

# vẽ cửa sổ 02
pen.penup()
pen.goto(-80, 0)
pen.pendown()
pen.right(90)
pen.color("black", "white")
pen.begin_fill()
for i in range(4):
    pen.forward(50)
    pen.left(90)
pen.end_fill()

# Vẽ đường ngang cửa sổ 
pen.penup()
pen.goto(-80, 25)
pen.pendown()
pen.color("black")
pen.forward(50)

# Vẽ đường dọc cửa sổ
pen.penup()
pen.goto(-55, 0)
pen.pendown()
pen.left(90)
pen.forward(50)

# Vẽ cửa chính
pen.penup()
pen.goto(-40, -97)
pen.pendown()
pen.right(90)
pen.color("red")
pen.begin_fill()
for i in range(2):
    pen.forward(50)
    pen.left(90)
    pen.forward(80)
    pen.left(90)
pen.end_fill()

# vẽ tay nắm cửa
pen.penup()
pen.goto(-30, -60)
pen.pendown()
pen.color("black")
pen.begin_fill()
pen.circle(5)
pen.end_fill()

turtle.done()
```
