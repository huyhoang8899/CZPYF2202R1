### Xử lý dữ liệu ngoại lai

```python
cat.head()
```

```python
sale.head()
```

```python
import seaborn as sns
```

```python
sale1=sale.groupby('Manufacturer')['Sumofsales'].sum().sort_values(ascending=False).to_frame()
sale1
```

```python
price=cat.groupby('Manufacturer')['Price_in_thousands'].mean().sort_values(ascending=False).to_frame()
price
```

```python
a = sale1.merge(price, left_on='Manufacturer', right_on='Manufacturer')
a
```

```python
sns.scatterplot(data=a, x='Price_in_thousands', y='Sumofsales')
```

```python
a.query("Sumofsales < 5000").sort_values(by='Sumofsales', ascending=False)
```

```python
sale2=sale.groupby('Model')['Sumofsales'].sum().sort_values(ascending=False).to_frame()
sale2
```

```python
price2=cat.groupby('Model')['Price_in_thousands'].mean().sort_values(ascending=False).to_frame()
price2
```

```python
b = sale2.merge(price2, left_on='Model', right_on='Model')
b
```

```python
sns.scatterplot(data=b, x='Price_in_thousands', y='Sumofsales')
```

### Phân tích sự ảnh hưởng của chỉ số kỹ thuật

```python
tong_hop = cat.merge(sale2, left_on='Model', right_on='Model')
tong_hop
```

```python
sns.pairplot(tong_hop)
```
