```python
# nhập 3 số a, b, c cho biết tổng của 3 số này chẵn hay lẻ

a = int(input("Nhập số a:"))
b = int(input("Nhập số b:"))
c = int(input("Nhập số c:"))
d = a + b + c

if (d % 2 == 0):
    print("Đây là số chẵn")
else:
    print("Đây là số lẻ")
```

```python
# nhập 6 số cho biết tổng của 6 số này chẵn hay lẻ
a = int(input("Nhập số a:"))
b = int(input("Nhập số b:"))
c = int(input("Nhập số c:"))
d = int(input("Nhập số d:"))
e = int(input("Nhập số e:"))
f = int(input("Nhập số f:"))
total = a + b + c + d + e + f

if (total % 2 == 0):
    print("Đây là số chẵn")
else:
    print("Đây là số lẻ")
```

## Cấu trúc lặp while

```python
i = 0
tong = 0

while i < 6:
    value = int(input("Nhập số: "))
    tong += value
    i += 1
    
print(f"Tổng của 6 số đã nhập: {tong}")
```

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
i = 0
n = int(input("nhập 1 số: "))
while i <= n:
    if i % 2 == 0:
        print(i)
        
    i += 1
```

### Ví dụ: Đảo ngược chuỗi

```python
content = input("Nhập chuỗi: ")
i = len(content) - 1

while i >= 0:
    print(content[i], end="")
    i -= 1
```

### Ví dụ: 
Nhập vào một số nguyên a

Cho S(k) = 1 + 2 + 3 + ... + k

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
#Bài toán S(k) = 1 + 2 + 3 + ... + k

a = int(input("Nhập a: "))
tong = 0
k = 0
while tong <= a:
    k = k+1
    tong = tong + k
print(f"Với k = {k} nhỏ nhất thì tổng = {tong} lớn hơn {a}")
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
# nhập vào số n, in ra 3 số lẻ đầu tiên trong khoảng từ n -- 1

n = int(input("Nhập n: "))
i = n
cont = 0

while i >= 1:
    if i % 2:
        print(i)
        cont += 1
    if cont == 3:
        break
        
    i -= 1
```

## Lệnh continue

```python
# in ra tất cả những số từ 1 --- 10 trừ các số 3,4,6

i = 0
while i < 10:
    i += 1
    
    if i == 3 or i == 4 or i ==6:
        continue
    
    print(i)
    
    
    
```
