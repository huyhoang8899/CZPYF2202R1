```python
from tkinter import *
from tkinter import messagebox

window = Tk()
window.title("Password Manager")

website_lbl = Label(window, text="Website")
username_lbl = Label(window, text="Email/Username")
password_lbl = Label(window, text = "Password")

website = Entry(window)
username = Entry(window)
password = Entry(window)

list_info = []

def clear_entry_value():
    website.delete(0,END)
    username.delete(0,END)
    password.delete(0,END)

def add_password():
    global list_info   # ch nh list_info bn ngoi  s dng trong hm
    print(f"Website: {website.get()}")
    print(f"Username: {username.get()}")
    print(f"Password: {password.get()}")
    info = (website.get(), username.get(), password.get())
    list_info.append(info)
    print(list_info)
    messagebox.showinfo("add password", "Password added successfully!")
    clear_entry_value()
    
def search_info():
    print(f"Looking on website: {website.get()}")
    for info in list_info:
        web_name = info[0]
        if web_name == website.get():
            username_text = info[1]
            password_text = info[2]
            messagebox.showinfo("Website information", f"Username: {username_text}, Password: {password_text}")
            break
    
    
search_btn = Button(window, text="Search", command=search_info)
add_btn = Button(window, text="Add", command=add_password)

website_lbl.grid(row=0,column=0) 
username_lbl.grid(row=1,column=0)
password_lbl.grid(row=2,column=0)

website.grid(row=0,column=1)
username.grid(row=1,column=1)
password.grid(row=2,column=1)

search_btn.grid(row=0,column=2)
add_btn.grid(row=3,column=1)

window.mainloop()
```
